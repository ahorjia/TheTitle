```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def cannyEdge(imgFileName):
    img = cv2.imread(imgFileName,0)
    edges = cv2.Canny(img,100,200)

    plt.figure(figsize = (20,50))
    
    plt.subplot(211)
    plt.imshow(img, cmap = 'gray')
    plt.title('Original Image')
    plt.xticks([])
    plt.yticks([])
    
    plt.subplot(212)
    plt.imshow(edges, cmap = 'gray')
    plt.title('Edge Image')
    plt.xticks([])
    plt.yticks([])

    plt.show()
```

```python
cannyEdge('pp1.jpg')
```

```python
cannyEdge('cd1.jpg')
```

```python
cannyEdge('titleBlock1.jpg')
```
